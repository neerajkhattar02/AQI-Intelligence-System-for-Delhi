```python
import pandas as pd

df = pd.read_csv(
    "../data/raw/india_aqi_raw.csv",
    encoding="latin1"
)

df.head()

df.info()

df.columns

df['location'].unique()

delhi_df = df[df['location'] == 'Delhi']
delhi_df.shape

delhi_df.head()
delhi_df.info()

cols_to_drop = [
    'stn_code',
    'agency',
    'type',
    'location_monitoring_station',
    'sampling_date'
]

delhi_df = delhi_df.drop(columns=cols_to_drop)

pollutant_cols = ['so2', 'no2', 'rspm', 'spm']

for col in pollutant_cols:
    delhi_df[col] = delhi_df[col].fillna(delhi_df[col].median())

delhi_df['date'] = pd.to_datetime(delhi_df['date'], errors='coerce')

delhi_df['year'] = delhi_df['date'].dt.year
delhi_df['month'] = delhi_df['date'].dt.month
delhi_df['day'] = delhi_df['date'].dt.day

delhi_df.to_csv("../data/processed/delhi_aqi_cleaned.csv", index=False)
print("Saved delhi_aqi_cleaned.csv")
```
