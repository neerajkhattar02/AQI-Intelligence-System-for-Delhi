```python
import pandas as pd

df = pd.read_csv("../data/processed/delhi_aqi_cleaned.csv")
df.head()

monthly_rspm = df.groupby("month")["rspm"].mean().reset_index()
monthly_rspm

def season_from_month(month):
    if month in [11, 12, 1]:
        return "Winter"
    elif month in [3, 4, 5]:
        return "Summer"
    elif month in [6, 7, 8]:
        return "Monsoon"
    else:
        return "Post-Monsoon"

df["season"] = df["month"].apply(season_from_month)

seasonal_rspm = df.groupby("season")["rspm"].mean().reset_index()
seasonal_rspm

def pollution_severity(rspm):
    if rspm < 100:
        return "Low"
    elif rspm < 200:
        return "Moderate"
    elif rspm < 300:
        return "High"
    else:
        return "Severe"

df["pollution_severity"] = df["rspm"].apply(pollution_severity)

df["pollution_severity"].value_counts()

severity_by_season = (
    df.groupby(["season", "pollution_severity"])
      .size()
      .reset_index(name="count")
)

severity_by_season

import joblib

rf = joblib.load("../models/rf_rspm_model.pkl")

example_input = pd.DataFrame([{
    "so2": 15,
    "no2": 40,
    "spm": 180,
    "year": 2024,
    "month": 12
}])

predicted_rspm = rf.predict(example_input)[0]
predicted_rspm

predicted_severity = pollution_severity(predicted_rspm)
predicted_severity
```
